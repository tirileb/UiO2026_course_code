```python
def move(frm,to):
  """Print the move from one peg to another."""
  print(f"Move disk from {frm} to {to}")

def move_via(frm,via,to):
  """Move via an auxiliary peg."""
  move(frm, via)
  move(via, to)

def hanoi_solver(n: int,source,auxiliary,target):
  """Print the moves to solve Tower of Hanoi with n disks."""
  if n == 1:
    move(source, target)
    return 1
  else:
    m1 = hanoi_solver(n - 1, source, target, auxiliary)
    move(source, target)
    m2 = hanoi_solver(n - 1, auxiliary, source, target)
    return m1 + 1 + m2 # return the number of moves

# run for various number of disks
for k in [3,4,7]:
  print(f'Solving Towers of Hanoi with {k} disks:')
  r = hanoi_solver(k,'A','B','C')
  print(f'With {k} disks the solution requires {r} moves\n')
```
